```python
__author__ = 'Aleksandar Skrbic'

import numpy as np
import cv2
import matplotlib
import matplotlib.pyplot as plt

PATH = '../../images/snowboarders.jpg'
```

```python
%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = 10, 8
```

```python
img = cv2.cvtColor(cv2.imread(PATH), cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, 'gray')
```

```python
blurred = cv2.GaussianBlur(img_gray, (11, 11), 0)
plt.imshow(blurred, 'gray')
```

```python
image_ada_bin = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 35, 15)
plt.imshow(image_ada_bin, 'gray')
```

```python
inverted = cv2.bitwise_not(image_ada_bin)
plt.imshow(inverted, 'gray')
```

```python
eroded = cv2.erode(inverted, (4,4), iterations=13)
plt.imshow(eroded, 'gray')
```

```python
dilated = cv2.dilate(eroded, (4,4), iterations=1)
plt.imshow(dilated, 'gray')
```

```python
img, contours, hierarchy = cv2.findContours(dilated, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

img = cv2.cvtColor(cv2.imread(PATH), cv2.COLOR_BGR2RGB)
cv2.drawContours(img, contours, -1, (255, 0, 0), 1)
print('Number of regions {}'.format(len(contours)))
plt.imshow(img)
```

```python
snowboarders = []
for contour in contours: 
    center, size, angle = cv2.minAreaRect(contour)
    width, height = size
    if width > 7 and width < 70:
        snowboarders.append(contour)

print("There is {} snowboarders on image.".format(len(snowboarders)))
img = cv2.cvtColor(cv2.imread(PATH), cv2.COLOR_BGR2RGB)
cv2.drawContours(img, snowboarders, -1, (255, 0, 0), 1)
plt.imshow(img)
```

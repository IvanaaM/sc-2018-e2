```python
__author__ = 'Aleksandar Skrbic'

import numpy as np
import cv2
import matplotlib
import matplotlib.pyplot as plt

PATH = '../../images/bulbs.jpg'
```

```python
%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = 16, 10
```

```python
img = cv2.cvtColor(cv2.imread(PATH), cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, 'gray')
```

```python
blurred = cv2.GaussianBlur(img_gray, (11, 11), 0)
plt.imshow(blurred, 'gray')
```

```python
image_bin = cv2.threshold(blurred, 234, 255, cv2.THRESH_BINARY)[1]
plt.imshow(image_bin, 'gray')
```

```python
image_bin = cv2.erode(image_bin, None, iterations=2)
image_bin = cv2.dilate(image_bin, None, iterations=4)
plt.imshow(image_bin, 'gray')
```

```python
img, contours, hierarchy = cv2.findContours(image_bin, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
print('There is {} bulbs on picture.'.format(len(contours)))

img = cv2.cvtColor(cv2.imread(PATH), cv2.COLOR_BGR2RGB)
cv2.drawContours(img, contours, -1, (255, 0, 0), 1)
plt.imshow(img)
```
